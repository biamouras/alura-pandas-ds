## Tratamento de dados faltantes

```python
import pandas as pd
```

```python
df = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
df.head()
```

Identificando valores null

```python
df.isnull()
```

```python
df.info()
# dá informação da quantidade de elementos não nulos de cada variável
```

Podemos ver que as variáveis 'Valor', 'Condomínio' e 'IPTU' tem valores nulos.

```python
df[df['Valor'].isnull()]
```

Como é a variável que mais importa, vamos remover as linhas que não tem o registro do valor de aluguel.

```python
df.dropna(subset = ['Valor'], inplace = True)
```

## Tratamento de condicional

O valor do condomínio também é importante para a análise. Então vamos avaliar os valores nulos desta variável.

```python
len(df[df['Condominio'].isnull()])
```

```python
selecao = (df['Tipo'] == 'Apartamento') & (df['Condominio'].isnull())
df[selecao]
```

Não faz sentido existir apartamentos que não tenham valor de condomínio. Então vamos eliminar esse caso.

```python
df = df[~selecao]
```

```python
len(df[df['Condominio'].isnull()])
```

Um caso ignorado na análise são as casas de condomínio. Vamos dar uma olhada nelas.

```python
selecao = (df['Tipo'] == 'Casa de Condomínio') & (df['Condominio'].isnull())
df[selecao]
```

Para evitar incompatibilidades com os exercícios futuros, não vou eliminar estes casos. Mas eles se enquadram na mesma questão que o apartamento sem condomínio.

Entende-se que os casos restantes no DataFrame que ainda são Null, na verdade, são valores 0. Então vamos substituí-los.

```python
df = df.fillna({'Condominio': 0, 'IPTU': 0})
df
```

```python
len(df[df['Condominio'].isnull()])
```

```python
# salvando em um novo arquivo
df.to_csv('dados/aluguel_residencial_limpo.csv', sep = ';')
```
