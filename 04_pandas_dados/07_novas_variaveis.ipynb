```python
import pandas as pd
```

## Criando novas variáveis

```python
dados = pd.read_csv('dados/aluguel_residencial_limpo.csv', sep = ';')
dados.head()
```

```python
dados['Valor_m2'] = dados['Valor'] / dados['Area']
dados
```

```python
# arredondando a variavel calculada
dados['Valor_m2'] = dados['Valor_m2'].round(2)
dados
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
casa
```

## Agregando variável

```python
# criando uma nova variável que agrega os tipos de imóveis em casa ou apto
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados
```

## Excluindo variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor_m2', 'Tipo', 'Suites']])
dados_aux
```

```python
del dados_aux['Tipo']
dados_aux
```
