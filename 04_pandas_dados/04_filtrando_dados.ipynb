Filtrando os imóveis residenciais para continuar com a análise

## Imóveis residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
# verificando os tipos de imóveis
list(dados['Tipo'].drop_duplicates())
```

```python
# determinando quais tipos são residencial - deixei comentado o que na aula não foi considerado como residencial
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 #'Flat',
 'Casa de Vila',
 #'Loft',
 #'Studio'
              ]
residencial
```

```python
selecao = dados['Tipo'].isin(residencial) # preparando a condição para filtrar
```

```python
dados_residencial = dados[selecao]
dados_residencial
```

```python
# verificando se a selecao funcionou
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
# atualizando o index
dados_residencial.index = range(len(dados_residencial))
dados_residencial
```

## Exportando a base de dados

```python
dados_residencial.to_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
# testando o arquivo salvo
df_residencial = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
df_residencial
# note que como quando salvamos não colocamos que não era para o índice ser exportado
```
