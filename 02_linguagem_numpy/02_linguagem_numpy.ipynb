## Condicionais - `if`

````python
# 1¬∫ item da lista - Nome do ve√≠culo
# 2¬∫ item da lista - Ano de fabrica√ß√£o
# 3¬∫ item da lista - Ve√≠culo √© zero km?

dados = [
    ['Jetta Variant', 2003, False],
    ['Passat', 1991, False],
    ['Crossfox', 1990, False],
    ['DS5', 2019, True],
    ['Aston Martin DB4', 2006, False],
    ['Palio Weekend', 2012, False],
    ['A5', 2019, True],
    ['S√©rie 3 Cabrio', 2009, False],
    ['Dodge Jorney', 2019, False],
    ['Carens', 2011, False]
]
dados
````

````python
# selecionando os carros 0 km

zero_km = []

for carro in dados:
  if carro[2]:
    zero_km.append(carro)

zero_km
````

````python
# selecionando carros usados

usados = []

for carro in dados:
  if carro[2] == False: 
    usados.append(carro)

usados
````

````python
# list comprehension
[carro for carro in dados if carro[2]]
````

## Condicionais - `if-else`

````python
zero_km, usados = [], []

for carro in dados:
  if carro[2]:
    zero_km.append(carro)
  else:
    usados.append(carro)

zero_km
````

````python
usados
````

### Exerc√≠cio

O IMC (√≠ndice de massa corporal) √© um indicador utilizado para detectar situa√ß√µes de obesidade ou desnutri√ß√£o. Para obter o IMC, basta calcular a raz√£o entre o peso de um indiv√≠duo (kg) e sua altura elevada ao quadrado, conforme mostra a f√≥rmula abaixo:

$IMC = \frac{Peso}{Altura^2}$

Considerando o enunciado acima, verifique cada alternativa e assinale as que calculam corretamente o IMC, n√£o apresentando erro(s) durante o seu c√°lculo.

````python
peso = np.array([106.0, 68.5, 75.0])
altura = np.array([1.9, 1.53, 1.75])
IMC = peso / (altura * altura)
IMC
````

````python
peso = np.array([106.0, 68.5, 75.0])
altura = np.array([1.9, 1.53, 1.75])
IMC = peso / altura ** 2
IMC
````
