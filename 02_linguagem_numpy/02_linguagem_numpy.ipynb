## Condicionais - `if`

````python
# 1º item da lista - Nome do veículo
# 2º item da lista - Ano de fabricação
# 3º item da lista - Veículo é zero km?

dados = [
    ['Jetta Variant', 2003, False],
    ['Passat', 1991, False],
    ['Crossfox', 1990, False],
    ['DS5', 2019, True],
    ['Aston Martin DB4', 2006, False],
    ['Palio Weekend', 2012, False],
    ['A5', 2019, True],
    ['Série 3 Cabrio', 2009, False],
    ['Dodge Jorney', 2019, False],
    ['Carens', 2011, False]
]
dados
````

````python
# selecionando os carros 0 km

zero_km = []

for carro in dados:
  if carro[2]:
    zero_km.append(carro)

zero_km
````

````python
# selecionando carros usados

usados = []

for carro in dados:
  if carro[2] == False: 
    usados.append(carro)

usados
````

````python
# list comprehension
[carro for carro in dados if carro[2]]
````

## Condicionais - `if-else`

````python
zero_km, usados = [], []

for carro in dados:
  if carro[2]:
    zero_km.append(carro)
  else:
    usados.append(carro)

zero_km
````

````python
usados
````

### Exercício

O IMC (índice de massa corporal) é um indicador utilizado para detectar situações de obesidade ou desnutrição. Para obter o IMC, basta calcular a razão entre o peso de um indivíduo (kg) e sua altura elevada ao quadrado, conforme mostra a fórmula abaixo:

$IMC = \frac{Peso}{Altura^2}$

Considerando o enunciado acima, verifique cada alternativa e assinale as que calculam corretamente o IMC, não apresentando erro(s) durante o seu cálculo.

````python
peso = np.array([106.0, 68.5, 75.0])
altura = np.array([1.9, 1.53, 1.75])
IMC = peso / (altura * altura)
IMC
````

````python
peso = np.array([106.0, 68.5, 75.0])
altura = np.array([1.9, 1.53, 1.75])
IMC = peso / altura ** 2
IMC
````
