## Operações com variáveis

```python
ano_atual = 2022
ano_fabricacao = 2013
km_total = 44410.0
km_media = km_total / (ano_atual - ano_fabricacao)
km_media
```

```python
km_total == km_total_2021
```

# Tipos e mudança de tipo de dados

A aula discutiu os tipos de dados, que podem ser:

 * `int`: números inteiros
 * `float`: números reais
 * `boolean`: boleanos (verdadeiro ou falso)
 * `str`: caracter

Para mudar o tipo de dado de uma variável, posso usar `int()`, `float()` ou `str()`.

```python
a, b, c, d = 10, 20, "Python é ", "legal"
```

```python
a + b
```

```python
c + d
```

Posso somar variáveis de tipos iguais, mas não de  tipos diferentes, sendo necessário mudar o tipo

## Formatando strings

São duas opções para formatar as `strings`. Usando `str.format()` ou `.f` 

```python
# sem definir variável, só aproveitando a ordem de inserção
print('Olá {}! Este é seu acesso {}'.format('Rodrigo', 32))
```

```python
# definindo o nome da variável
print('Olá {nome}! Este é seu acesso {acesso}'.format(nome = 'Rodrigo', acesso = 32))
```

```python
# colocando a variável antes
nome = 'Rodrigo'
acesso = 32
print(f'Olá {nome}! Este é seu acesso {acesso}')
```
